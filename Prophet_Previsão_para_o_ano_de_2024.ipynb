```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Recarregar os dados fornecidos para análise (ajustado para previsão de 2024)
data = {
    "Year": list(range(2000, 2024)),
    "Bragantina": [44440, 7031, 5646, 7394, 9906, 8728, 10512, 9448, 8003, 4943, 4879, 3071, 2717, 2868, 2868, 2955, 3177, 695, 2163, 3600, 3581, 3669, 3803, 3816],
    "Salgado": [34464, 4254, 4724, 7114, 8102, 8071, 15875, 11323, 7807, 8931, 7766, 7672, 7672, 3828, 3828, 5433, 4926, None, 430, 1250, 1250, 1250, 1300, 1668],
    "Guamá": [26052, 9103, 12086, 8562, 7551, 7692, 7554, 7893, 5761, 5771, 4916, 5629, 5652, 4152, 4440, 11820, 1682, 1410, 1144, 1572, 1687, 1779, 1697, 1592],
    "Tomé-Açu": [6608, 1914, 1932, 2363, 2379, 3920, 4712, 5220, 4601, 1781, 1684, 1924, 1924, 1569, 1706, 2471, 1541, 2150, 3088, 2176, 3031, 2353, 2360, 2155],
    "Cametá": [1195, 132, 216, 705, 3600, 10969, 1634, 1114, 244, 244, 204, 254, 120, 120, 120, 120, 120, 360, 120, 120, 300, 200, 220, 213]
}

# Criar DataFrame
df = pd.DataFrame(data)

# Lista de regiões para previsão de 2024
regions = ["Bragantina", "Salgado", "Guamá", "Tomé-Açu", "Cametá"]

# Função para aplicar Prophet e calcular métricas de erro
def train_and_evaluate_prophet_model(df, regions):
    results = []

    for region in regions:
        df_region = df[["Year", region]].dropna()
        df_prophet = pd.DataFrame({"ds": pd.to_datetime(df_region["Year"], format="%Y"), "y": df_region[region]})

        # Aplicar transformação logarítmica para evitar valores negativos
        df_prophet["y"] = np.log1p(df_prophet["y"])

        model = Prophet()
        model.fit(df_prophet)

        future = pd.DataFrame({"ds": pd.to_datetime(["2024-01-01"])})
        forecast = model.predict(future)
        predicted_value = np.expm1(forecast["yhat"].iloc[0])  # Reverter a transformação logarítmica
        predicted_value = max(predicted_value, 0)  # Garantir que não seja negativo

        # Calcular métricas de erro
        future_full = model.make_future_dataframe(periods=1, freq='Y')
        forecast_full = model.predict(future_full)
        y_pred = np.expm1(forecast_full["yhat"][:-1])  # Remover o valor previsto para 2024
        y_real = np.expm1(df_prophet["y"])

        mae = mean_absolute_error(y_real, y_pred)
        mse = mean_squared_error(y_real, y_pred)
        rmse = np.sqrt(mse)

        results.append({
            "Região": region,
            "MAE": mae,
            "MSE": mse,
            "RMSE": rmse,
            "Produção Estimada em 2024": predicted_value
        })

        # Gerar gráfico para a previsão
        plt.figure(figsize=(8, 5))
        plt.plot(df_prophet["ds"], y_real, color="blue", label="Dados reais")
        plt.plot(forecast_full["ds"], np.expm1(forecast_full["yhat"]), color="red", linestyle="dashed", label="Previsão Prophet")
        plt.scatter(pd.to_datetime("2024-01-01"), predicted_value, color="green", marker="o", label="Previsão 2024")
        plt.title(f"Previsão Prophet Ajustada - {region}")
        plt.xlabel("Ano")
        plt.ylabel("Produção")
        plt.legend()
        plt.grid()
        plt.show()

    return pd.DataFrame(results)

# Executar modelo Prophet para previsão de 2024
results_prophet_2024 = train_and_evaluate_prophet_model(df, regions)

# Exibir tabela com previsões e métricas
from IPython.display import display
print("\nResultados das previsões para 2024 com Prophet:")
display(results_prophet_2024)
```

```python
# Importação das bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split

# Recarregar os dados fornecidos para análise (ajustado para previsão de 2023)
data = {
    "Year": list(range(2000, 2023)),
    "Bragantina": [44440, 7031, 5646, 7394, 9906, 8728, 10512, 9448, 8003, 4943, 4879, 3071, 2717, 2868, 2868, 2955, 3177, 695, 2163, 3600, 3581, 3669, 3803],
    "Salgado": [34464, 4254, 4724, 7114, 8102, 8071, 15875, 11323, 7807, 8931, 7766, 7672, 7672, 3828, 3828, 5433, 4926, None, 430, 1250, 1250, 1250, 1300],
    "Guamá": [26052, 9103, 12086, 8562, 7551, 7692, 7554, 7893, 5761, 5771, 4916, 5629, 5652, 4152, 4440, 11820, 1682, 1410, 1144, 1572, 1687, 1779, 1697],
    "Tomé-Açu": [6608, 1914, 1932, 2363, 2379, 3920, 4712, 5220, 4601, 1781, 1684, 1924, 1924, 1569, 1706, 2471, 1541, 2150, 3088, 2176, 3031, 2353, 2360],
    "Cametá": [1195, 132, 216, 705, 3600, 10969, 1634, 1114, 244, 244, 204, 254, 120, 120, 120, 120, 120, 360, 120, 120, 300, 200, 220]
}

# Criar DataFrame
df = pd.DataFrame(data)

# Lista de regiões para previsão de 2023
regions = ["Bragantina", "Salgado", "Guamá", "Tomé-Açu", "Cametá"]

# Função para aplicar Prophet com divisão de dados e gerar gráficos
def train_evaluate_and_plot_prophet(df, regions):
    results = []

    for region in regions:
        df_region = df[["Year", region]].dropna()
        df_prophet = pd.DataFrame({"ds": pd.to_datetime(df_region["Year"], format="%Y"), "y": df_region[region]})

        # Aplicar transformação logarítmica para evitar valores negativos
        df_prophet["y"] = np.log1p(df_prophet["y"])

        # Dividir os dados entre treino (80%) e teste (20%)
        train, test = train_test_split(df_prophet, test_size=0.2, shuffle=False)

        # Treinar modelo Prophet
        model = Prophet()
        model.fit(train)

        # Previsões para o conjunto de teste
        future_test = pd.DataFrame({"ds": test["ds"]})
        forecast_test = model.predict(future_test)
        y_pred_test = np.expm1(forecast_test["yhat"])
        y_real_test = np.expm1(test["y"])

        # Calcular métricas de erro no conjunto de teste
        mae = mean_absolute_error(y_real_test, y_pred_test)
        mse = mean_squared_error(y_real_test, y_pred_test)
        rmse = np.sqrt(mse)

        # Previsão para 2023
        future = pd.DataFrame({"ds": pd.to_datetime(["2023-01-01"])})
        forecast = model.predict(future)
        predicted_value = np.expm1(forecast["yhat"].iloc[0])  # Reverter transformação logarítmica
        predicted_value = max(predicted_value, 0)  # Garantir que não seja negativo

        results.append({
            "Região": region,
            "MAE": mae,
            "MSE": mse,
            "RMSE": rmse,
            "Produção Estimada em 2023": predicted_value
        })

        # Gerar gráfico para a previsão
        future_full = model.make_future_dataframe(periods=1, freq='Y')
        forecast_full = model.predict(future_full)
        y_pred_full = np.expm1(forecast_full["yhat"])
        y_real = np.expm1(df_prophet["y"])

        plt.figure(figsize=(8, 5))
        plt.plot(df_prophet["ds"], y_real, color="blue", label="Dados reais")
        plt.plot(forecast_full["ds"], y_pred_full, color="red", linestyle="dashed", label="Previsão Prophet")
        plt.scatter(pd.to_datetime("2023-01-01"), y_pred_full.iloc[-1], color="green", marker="o", label="Previsão 2023")
        plt.title(f"Previsão Prophet Ajustada - {region}")
        plt.xlabel("Ano")
        plt.ylabel("Produção")
        plt.legend()
        plt.grid()
        plt.show()

    return pd.DataFrame(results)

# Executar modelo Prophet com divisão de dados e gerar gráficos
results_prophet_with_graphs = train_evaluate_and_plot_prophet(df, regions)


# Exibir tabela com previsões e métricas
from IPython.display import display
print("\nResultados das previsões para 2023 com Prophet:")
display(results_prophet_with_graphs)
```
